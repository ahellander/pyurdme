# Chemical Potential in 2D 
#### One diffusing molecule diffuses around identical crowders.
RCrowder (R) = radius of crowders

RDiff (r) = radius of diffusing molecule

$\phi$ = percentage of occupied volume

h = voxel size in mesoscopic model

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
Crowding = np.arange(0,0.7,0.1)
def ChemicalPotential2D(R,r,h,Crowding):
    
    V = h*h
    N = np.ceil(Crowding*V/(R*R*np.pi))
    
    xi0 = np.pi*N/(4*V)
    xi1 = np.pi*N*R/(2*V)
    xi2 = np.pi*N*R**2/V
    
    x = -np.log(1-xi2)+4*xi1*r/(1-xi2)+r**2*(4*xi0/(1-xi2)+4*xi1**2/((1-xi2)**2))
    
    gamma = np.exp(x)
    
    return gamma
```

### RCrowder = 5e-2, RDiff = 0

```python
gamma1 = ChemicalPotential2D(5e-2,0,1,Crowding)
gamma2 = ChemicalPotential2D(1e-1,0,1,Crowding)

plt.plot(Crowding,1/gamma1,'b', label='R=5e-2')
plt.plot(Crowding,1/gamma2,'r', label='R=1e-1')
plt.plot(Crowding,1-Crowding,'k--', label='$1-\phi$')
plt.legend(loc='lower left')
plt.xlabel('$\phi$')
plt.ylabel('$d_{ij}/d_{0ij}$')
plt.show()
```
