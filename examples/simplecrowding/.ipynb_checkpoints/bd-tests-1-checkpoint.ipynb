### Test 1 - Irreversible binary reaction - radial probability distribution

In the first test we compare our simulation results with the analitical formula that calculates the radial probability distribution for two particles that diffuse and react with a given rate (see Kim and Shin - _Physical Review Letters_, 82(7), 1578-1581 (1999) - Eq. 17).

The following cell defines functions used for test $1$.

```python
def a_irr(k_r, s, D):
    """
    Equation 17.
    s : sigma, distance at which a reaction can occour.
    D : diffusion rate.
    k_D : diffusion limited reaction rate. = 4 * pi * s * D
    k_r : intrinsic reaction rate for two particles within a reaction distance s.
    """
    k_D = 4.0 * pi * s * D
    return (1 + k_r/k_D) * sqrt(D) / s

def W(a,b):
    """
    W(a,b) = exp(2ab + b^2) * erfc(a+b).
    For numerical stability we rearrange:
        exp(2ab+b^2) * erfc(a+b) = exp( (a+b)^2 ) * exp(-a^2) * erfc(a+b)= exp(-a^2) * erfce( a+b )
    Then we can use erfce() to avoid numerical over/underflow issues
    which occour as a+b gets large.
    Consider for example what happens when a = 1.75 and b = 24.85.
    Then exp(2ab+b^2) = 1.61712462288e+306
    and erfc(a+b) = 5.92210224519e-310
    and we loose precision in the multiplication of these two terms.
    """
    #  instead of 'return exp(2.0*a*b + b*b) * erfc(a+b)' we then use
    return erfce( a+b ) * exp(-a*a)

def p_irr(r, t, r_0, k_r, sigma, D):
    
    A = 1.0/sqrt(4*pi*t) * (\
    exp( - (r-r_0)*(r-r_0)/(4*D*t) ) + \
    exp( - (r + r_0 - 2*sigma) * (r + r_0 - 2*sigma) /(4*D*t) ) \
    )

    a_ = a_irr(k_r, sigma, D)
    w_ = W( (r + r_0 - 2*sigma)/sqrt(4*D*t), a_ * sqrt(t) )
    B = a_ * w_
    return (A - B) / ( 4 * pi * r * r_0 * sqrt(D) )
```

#### Simulation details

Two particles $A$ and $B$ are modeled as hard spheres and are initially placed at a distance $r_0$ equal to the sum of their radii $R_A=0.5$ and $R_B=0.5$. Each simulation runs for a time $t_{sim}=20$. Four independent simulation sets have been completed using four different time steps ($\Delta t$): $1.0$ $0.1$ $0.01$ and $0.001$, respectively.
The two particles react irreversibly following: $A+B \rightarrow C$, with a reaction rate $k_r = 0.16k_D$ where $k_D=4 \pi \sigma D$ with $\sigma=R_A+R_B$. For each of the four time steps we run $100000$ simulations. At the beginning of each simulation, after the first particle is placed, the polar and azimuthal angles for the second particle are chosen randomly. If a reaction happens before the end we discard the entire run. If the two particles did not react at the end of the simulation we calculate and save their final distance. The goal of this simulation protocol is to generate enough data in order to calculate the radial probability distribution for the $A-B$ distance. All quantities are expressed in arbitrary units. The following cell displays the input data used in our simulations.

```python
D=1 # D = D1 + D2 where D1 = D2 = 0.5 are the diffusion coefficients of the two particles
sigma=1 # reaction distance, for hard spheres this is simply the sum of the particles' radii
r0=1 # distance at time $t=0$
k_D = 4 * pi * sigma * D # diffusion rate
k_r=0.16*k_D # reaction rate
t=20 # simulation time
```

```python
pippo = []
r = []
for i in range(1,1000):
    dr = (1+(i-1)*30.0/1000.0)
    r.append(dr)
    pippo.append(4.0*pi*dr*dr*p_irr(dr, t, r0, k_r, sigma, D))
```

The data files storing the final distances for the four different time steps (see above) have self-explanatory names.

```python
f = open('distances001.out', 'r')
datas = []
for line in f.readlines():
    row = line.strip().split()
    datas.append(row)

f.close()
    
data = np.array(datas,np.float64)

rlen = round(np.max(r)-np.min(r))
#nb=int(rlen / (3.5 * np.std(data) / pow(len(data),1.0/3.0))) #Scott
nb=int(rlen / (2.0 * (np.percentile(data,75) - np.percentile(data,25)) / pow(len(data),1.0/3.0))) #Freedman-Diaconis
#nb = int(sqrt(len(data))) #square rule
#nb = int((2.0 * pow(len(data),1.0/3.0))) # Rice
nnn = len(data)/100000.0
hist, bins = np.histogram(data,bins=nb,range=(r0,30),density=True)
widths = np.diff(bins)
hist*=nnn #normalize against the number of "survivors"

plt.rc("font", size=14)
plt.figure(figsize=(10.5, 6))
#plt.bar(bins[:-1],hist,widths)
plt.plot(bins[:-1]+0.5*(30.0-r0)/(len(bins)),hist,'or',markersize=10,mfc='none',label='BD')
plt.plot(r,pippo,'b-',label='Theory')
legend()
plt.xlabel('distance',fontsize=18)
plt.ylabel('radial prob. distr.',fontsize=18)
plt.text(15,0.07,'$\Delta t = 0.01$',fontsize=18,bbox=dict(facecolor='red', alpha=0.1))
plt.savefig("irr.eps",dpi = 300)
plt.savefig("irr.pdf",dpi = 300)
plt.show()
```

The plot shows the simulation data (red circles) and the analytical formula (blue line). The simulation results shown in the plot are obtained using $\Delta t = 0.01$. Analogous results are obtained with $\Delta t = 0.001$. Results from the simulations where $\Delta t = 0.1$ and $\Delta t = 1.0$ show larger differences with respect to the analytical curve.

### Test 2 - Annihilation

In the second test we consider a system of identical particles reacting upon collision and annihilating each other. We compare our simulation results with an analytical formula for the decrease of the particle concentration over time in such a system (see Barenburg et al. - _The Journal of Chemical Physics_, 117, 9202-9214 (2002) - Eq. 24).

#### Simulation details

$N=10^5$ particles are modeled as hard spheres and are randomly placed in a periodic cubic box of volume $V=10^7$.
Two particles react and annihilate each other on contact. Four independent simulations have been completed using four different time steps ($\Delta t$): $1.0$ $0.1$ $0.01$ and $0.001$, respectively. Each simulation ran for a time $t_{sim}=350$.  The particle's radius is $R=1.0$ and the diffusion coefficient is $D=1.0$. The input data used in our simulations is the same we used in the first test. In the following cell we calculate the time evolution of the number of particles using the analytical formula cited above. All quantities are expressed in arbitrary units.

```python
t = []
pippo2 = []
npart = 100000.0 # number of particles
vol = 10000000.0 # periodic box volume
c_0 = npart/vol # number concentration
k_D = 4 * pi * sigma * D # diffusion-limited reaction rate
for i in range(0,350):
    t.append(i)
    pippo2.append(vol/(2.0*k_D*i + (1.0/c_0)))    
```

The data files storing the final distances for the four different time steps (see above) have self-explanatory names.

```python
f2 = open('annihilation-ts001.out', 'r')
datas2 = []
for line in f2.readlines():
    row2 = line.strip().split()
    datas2.append(row2)

f2.close()    
data2 = np.array(datas2,np.int32)

plt.rc("font", size=14)
plt.figure(figsize=(10.5, 6))
plt.semilogy(t[::10],data2[::10],'ro', markersize=10, mfc='none',label='BD')
plt.semilogy(t,pippo2,'b-',label='Theory')
legend()
plt.xlabel('time',fontsize=18)
plt.ylabel('number of particles',fontsize=18)
plt.text(60,10000,'$\Delta t = 0.01$',fontsize=18,bbox=dict(facecolor='red', alpha=0.1))
plt.savefig("ann.eps",dpi = 300)
plt.savefig("ann.pdf",dpi = 300)
plt.show()
```

The plot shows the simulation data (red circles) and the analytical formula (blue line). The simulation results shown in the plot are obtained using $\Delta t = 0.01$. Analogous results are obtained with $\Delta t = 0.001$. Results from the simulations where $\Delta t = 0.1$ and $\Delta t = 1.0$ show larger differences with respect to the analytical curve.

### Test 3 - Diffusion under crowding conditions

In the third test we calculate the diffusion coefficient under different particle densities (concentrations) and observe the expected deviation from ideal solution conditions (infinite diluition). A comparison is made with a formula obtained from molecular dynamics simulations (see Speedy - _Molecular Physics_, 62, 509 (1987) and for a more complete computational analysis of diffusion in hard-sphere systems, Heyes - _The Journal of Physical Chemistry B_ 111(6), 1455-1464 (2007)).

#### Simulations details

$N=512$ particles are modeled as non-reacting hard spheres and are randomly placed in a periodic cubic box.
The volume of the cubic box is calculated as $V=(2R)^3N/\rho$ where $R=1.0$ is the hard sphere radius and $\rho$ is the normalized number density. We ran $6$ independent simulations, varying the normalized number density $\rho$. In detail, $\rho$ takes $6$ different values in the interval $[0.001,0.5]$. The diffusion coefficient is $D=0.5$ (ideal solution conditions), the timestep is $\Delta t = 0.01$ and the simulation time is $t_{sim}=200$. All quantities are expressed in arbitrary units. The diffusion coefficient is calculated from the mean square displacement of the particles using a maximum time window of $0.25t_{sim}$ and utilizing only the second half of the mean square displacement values.

```python
def dhs(rho,sigma,D):
    return D*(1-(rho/1.09))*(1+rho*rho*(0.4-0.83*rho*rho))

dhs(0.25,1,0.5)
```

```python
f3 = open('Diffusion-coeff-new.dat', 'r')
datas3 = []
for line in f3.readlines():
    row3 = line.strip().split()
    datas3.append(row3)

f3.close()    
    
data3 = np.array(datas3,np.float64)
rho = data3[:,0]
diffuse = data3[:,1]
#sdev = data3[:,2]
speedy = data3[:,3]

plt.rc("font", size=14)
plt.figure(figsize=(10.5, 6))
plt.plot(rho,diffuse,'ro', mfc='none',markersize=10, label='BD')
plt.plot(rho,speedy,'bo',markersize=10 ,label='Speedy 1987')
plt.axhline(y=0.5, xmin=0, xmax=1, label='zero-density limit')
legend()
plt.xlabel('normalized number density',fontsize=18)
plt.ylabel('diffusion coefficient',fontsize=18)
plt.axis([-0.05,0.65,-0.05,0.65 ])
plt.text(0.2,0.25,'$\Delta t = 0.01$',fontsize=18,bbox=dict(facecolor='red', alpha=0.1))
plt.savefig("diff.eps",dpi = 300)
plt.savefig("diff.pdf",dpi = 300)
plt.show()
```
